```python
# 1. CALCULATE AREA OF CIRCLE

def area_of_circle():
    radius = input ("Enter radius of the circle: ")
    try:
        value = float(radius)
        area_of_circle = round((math.pi) * (math.pow(value,2)), 10)
        print("Area of Circle : ", area_of_circle)
    except ValueError:
        print("Please input correct radius of circle !!")

# 2. CHECKING POSTIVE, NEGATIVE INTEGERS AND ZERO

def numbers_check():
    number = input ("Enter any positive and negative number: ")
    try:
        value = int(number)
        if value < 0:
            print("'" ,value ,"'" , "Negative Number Entered !!")
        elif value > 0:
            print("'" ,value ,"'" , "Positive Number Entered !!")
        else :
            print("'" ,value ,"'" , "Zero Entered !!")
    except ValueError:
        print("That's not a number !!")

# 3. CHECKING THE TWO NUMBERS ARE COMPLETELY DIVISIBLE OR NOT

def divisiblity_check():
    value = input ("Enter numerator: ")
    value2 = input ("Enter denominator: ")
    try:
        numerator = int(value)
        denominator = int(value2)
        divide_result = numerator % denominator
        if divide_result == 0:
            print("Number " ,  numerator , " is completely divisible by " , denominator)
        else:
            print("Number " , numerator , " is not completely divisible by " , denominator)
    except ValueError:
        print("That's not a number !!")

# 4. CALCULATE NUMBER OF DAYS BETWEEN TWO DATES


def numOfDays():
    dd1, mm1, yy1 = input("Enter a date in (dd mm yy) format: ").split()
    dd2, mm2, yy2 = input("Enter a date in (dd mm yy) format: ").split()
    try:
        date1 = date(int(yy1), int(mm1), int(dd1))
        date2 = date(int(yy2), int(mm2), int(dd2))
        timespan = (date2 - date1).days
        print("There are",timespan,"days in between", dd1,"/",mm1,"/",yy1,"and", dd2,"/",mm2,"/",yy2)
    except ValueError:
        print("That's not a correct date !!")


# 5. CALCULATE VOLUME OF SPHERE

def calculate_volume():
    radius = input ("Enter radius of sphere: ")
    try:
        value = float(radius)
        volume = round(((math.pi) * math.pow(value,2) * 4/3), 2)
        print("Volume of sphere: " , volume)

    except ValueError:
        print("That's not a number !!")


# 6. COPY STRINGS N TIMES

def copy_strings():
    message = input ("Enter message: ")
    number = input ("How many copies of string you need: ")
    try:
        value = int(number)
        if value <= 0:
            print("That's not correct number: ")
        else:
            print(value, "Copies of",message," are" , (message * value))

    except ValueError:
        print("That's not a number !!")

# 7. EVEN OR ODD NUMBER

def even_or_odd():
    number = input ("Enter Number: ")
    try:
        value = int(number)
        if value <= 0:
            print("That's not correct number: ")
        else:
            if value % 2 == 0:
                print(value,"is Even")
            else:
                print(value,"is Odd")

    except ValueError:
        print("That's not a number !!")

# 8. VOWEL TESTER

def vowel_tester():
    char = input("Input a letter of the alphabet: ")
    if char in ['a', 'e', 'i', 'o', 'u'] or char in ['A', 'E', 'I', 'O', 'U']:
	    print(char, "is a vowel.")
    elif char == "y" or char == "Y":
	    print("Sometimes letter y stand for vowel, sometimes stand for consonant.")
    else:
	    print(char, "is a consonant.")

# 9. TRIANGLE AREA

def area_triangle():
    base = input("Enter magnitude of Triangle base: ")
    height = input("Enter Magnitude of Triangle Height: ")
    try:
        value_base = float(base)
        value_height = float(height)
        area = 1/2 * value_base * value_height
        print("Area of a Triangle with Height",height,"and Base",base,"is",int(area))
    except ValueError:
        print("That's not a number !!")

# 10. CALCULATE INTEREST

def interest():
    principal_amount = input("Please enter principal amount: ")
    rate_of_interest = input("Please Enter Rate of interest in %: ")
    years =  input("Enter number of years for investment: ")

    try:
        value = int(principal_amount)
        value1 = float(rate_of_interest)
        value2 = int(years)
        if value <= 0:
            print("That's not correct number: ")
        elif value1 <= 0:
            print("That's not correct number: ")
        elif value2 <= 0:
            print("That's not correct number: ")
        else:
            interest = (((value1/100) * value) * value2)
            print(interest)
            total_amount = interest + value
            print("After", value2 ,"years your principal amount", value ,"over an interest rate of", value1 ,"% will be",total_amount)


    except ValueError:
        print("That's not a number !!")

# 11. EUCLIDEAN DISTANCE

def distance():
    x1 = input("Enter Co-ordinate for x1: ")
    x2 = input("Enter Co-ordinate for x2: ")
    y1 = input("Enter Co-ordinate for y1: ")
    y2=  input("Enter Co-ordinate for y2: ")

    try:
        value_x1 = int(x1)
        value_x2 = int(x2)
        value_y1 = int(y1)
        value_y2 = int(y2)
        distance = math.sqrt(math.pow((value_x1 - value_y1),2)+math.pow((value_x2 - value_y2),2))
        print("Distance between points (",value_x1,", ",value_y1,") and (",value_x2,", ",value_y2,") is",distance)

    except ValueError:
        print("That's not a number !!")


# 12. FEET TO CENTIMETER CONVERTER

def feet_to_centimeter():
    feet = input("Enter Height in Feet: ")
    try:
        value = float(feet)
        cm = value * 30.48
        print("There are",cm,"cm in",feet,"ft")

    except ValueError:
        print("That's not a number !!")

# 13. BMI CALCULATOR

def bmi_calculate():
    height = input("Enter Height in Cm: ")
    weight = input("Enter Weight in Kg: ")
    try:
        value_height = float(height) / 100
        value_weight = float(weight)
        bmi = round(value_weight / math.pow(value_height, 2), 2)
        print("Your BMI is", bmi)

    except ValueError:
        print("That's not a number !!")

# 14. SUM OF N POSITIVE INTEGERS

def sum_of_integers():
    limit = input("Enter value of n: ")
    try:
        value = int(limit)
        sum = int((value * (value + 1)) / 2)
        print("Sum of n Positive integers till",value,"is",sum)
    except ValueError:
        print("That's not a number !!")

# 15. DIGITS SUM OF A NUMBER

def sum_of_numbers():
    number = input("Enter a number: ")
    try:
        value = int(number)
        x  = value // 10
        x1 = (value - x*10) //1
        print("The sum of digits in the number is", x+x1)
    except ValueError:
        print("That's not a number !!")

# 16. DECIMAL TO BINARY CONVERTER

def decimal_to_binary():
    decimal_number = input("Enter a decimal number: ")
    try:
        value = int(decimal_number)
        if value > 1:
            print("Binary Representation of 5 is", bin(value).replace("0b", ""))
        else:
            print("Please input a correct number !!")

    except ValueError:
        print("That's not a number !!")


# 17. BINARY TO DECIMAL CONVERTER

def binary_to_decimal():
    binary_num = list(input("Input a binary number: "))
    value = 0
    for i in range(len(binary_num)):
        digit = binary_num.pop()
        if digit == '1':
            value = value + pow(2, i)
    print("The decimal value of the number is", value)

# 18. VOWEL AND CONSONANTS COUNTER

def vowel_counter():
    text = list(input("Enter text: "))
    vowel_count = 0
    consonants_count = 0
    space = 0
    for i in range(len(text)):
        alphabet = text.pop()
        if alphabet in ['a', 'e', 'i', 'o', 'u'] or alphabet in ['A', 'E', 'I', 'O', 'U']:
	        vowel_count = vowel_count + 1
        elif alphabet == " ":
            space = space + 1
        else:
	        consonants_count = consonants_count + 1
    print("Vowels: ",vowel_count)
    print("Consonants: ",consonants_count)

# 19. PALINDROME TESTER

def test_palindrome():
    text = input("Enter text: ")
    text = text.casefold()
    rev_text = reversed(text)
    if list(text) == list(rev_text):
        print("Text",text,"is Palindrome")
    else:
        print("Text",text,"is not a Palindrome")

# 20. ALPHABETS AND NUMBERS COUNTER

def digits_counter():
    numbers_count = 0
    alphabets_count = 0
    space_count = 0
    special_count = 0
    text = input("Enter text: ")
    for char in text:
        if char.isdigit():
	        numbers_count = numbers_count + 1
        elif char.isalpha():
            alphabets_count = alphabets_count + 1
        elif char.isspace():
	        space_count = space_count + 1
        else:
            special_count = special_count + 1
    print("Numbers: ",numbers_count)
    print("Alphabets: ",alphabets_count)
    print("Special Characters: ",special_count)
    print("Spaces: ",space_count)

#   PRINTING PATTERNS IN PYTHON

# 21. PATTERN 1

def  pattern_1():

    char = '*'
    n = 5
    for i in range(0, 5):
        for j in range(0, i + 1):
            print("*", end=" ")
        print("\r")
    for k in range(5, 0, -1):
        for l in range(0, k - 1):
            print("*", end=" ")
        print("\r")

# 22. PATTERN 2

def pattern_2():
    lastNumber = 6
    for row in range(1, lastNumber):
        for column in range(1, row + 1):
            print(column, end=' ')
        print("")
    for row in range(lastNumber, 1, -1):
        for column in range(1, row - 1):
            print(column, end=' ')
        print("")

# 23. PATTERN 3

def pattern_3():
    for num in range(10):
        for i in range(num):
            print (num, end=" ") #print number
        print("\r")


def functions():
    try:
        print("---------------- ASSIGNMENT NUMBER 1 ---------------- ")
        print(" 1.Calculate Area of a Circle \n 2.Check Number either positive, negative or zero \n 3.Divisibility Check of two numbers \n 4.Days Calculator \n 5.Calculate Volume of a sphere \n 6.Copy string n times \n 7.Check if number is Even or Odd \n 8.Vowel Tester \n 9.Triangle area \n 10.Calculate Interest \n 11.Euclidean distance \n 12.Feet to Centimeter Converter \n 13.BMI Calculator \n 14.Sum of n Positive Integers \n 15.Digits Sum of a Number \n 16.Decimal to Binary Converter \n 17.Binary to Decimal Converter \n 18.Vowel and Consonants Counter \n 19.Palindrome tester \n 20.Count Alphabets, Numbers and Special Characters \n 21.Write a Python program to construct the triangle(*) pattern \n 22.Write a Python program to construct the numbers pattern \n 23.Write a Python program to construct the numbers (1 to 9) pattern \n")
        argument = input("Enter your choice: ")
        input_func = int(argument)
        if input_func == 1:
            area_of_circle()
        if input_func == 2:
            numbers_check()
        if input_func == 3:
            divisiblity_check()
        if input_func == 4:
            numOfDays()
        if input_func == 5:
            calculate_volume()
        if input_func == 6:
            copy_strings()
        if input_func == 7:
            even_or_odd()
        if input_func == 8:
            vowel_tester()
        if input_func == 9:
            area_triangle()
        if input_func == 10:
            interest()
        if input_func == 11:
            distance()
        if input_func == 12:
            feet_to_centimeter()
        if input_func == 13:
            bmi_calculate()
        if input_func == 14:
            sum_of_integers()
        if input_func == 15:
            sum_of_numbers()
        if input_func == 16:
            decimal_to_binary()
        if input_func == 17:
            binary_to_decimal()
        if input_func == 18:
            vowel_counter()
        if input_func == 19:
            test_palindrome()
        if input_func == 20:
            digits_counter()
        if input_func == 21:
            pattern_1()
        if input_func == 22:
            pattern_2()
        if input_func == 23:
            pattern_3()
    except ValueError:
        print("Please input correct function index !!")

functions()
---------------- ASSIGNMENT NUMBER 1 ---------------- 
 1.Calculate Area of a Circle 
 2.Check Number either positive, negative or zero 
 3.Divisibility Check of two numbers 
 4.Days Calculator 
 5.Calculate Volume of a sphere 
 6.Copy string n times 
 7.Check if number is Even or Odd 
 8.Vowel Tester 
 9.Triangle area 
 10.Calculate Interest 
 11.Euclidean distance 
 12.Feet to Centimeter Converter 
 13.BMI Calculator 
 14.Sum of n Positive Integers 
 15.Digits Sum of a Number 
 16.Decimal to Binary Converter 
 17.Binary to Decimal Converter 
 18.Vowel and Consonants Counter 
 19.Palindrome tester 
 20.Count Alphabets, Numbers and Special Characters 
 21.Write a Python program to construct the triangle(*) pattern 
 22.Write a Python program to construct the numbers pattern 
 23.Write a Python program to construct the numbers (1 to 9) pattern 

Enter your choice: 1
Enter radius of the circle: 10
Area of Circle :  314.159265359
```
